```python
from keras.models import load_model
from vis.visualization import visualize_cam, visualize_saliency
import argparse
import cv2
from keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# model_dir = '../male_female_cnn/santa_not_santa.model'
model_dir = 'da_last4_layers.h5'
image_dir = '../male_female_cnn/examples/dave.jpeg'

model = load_model(model_dir)
image = cv2.imread(image_dir)
original_image = image.copy()
```

```python
conv_layers = []
layers_by_id = {}

for idx, _layer in enumerate(model.layers[0].layers):
    layers_by_id[idx] = _layer.name
    if "conv" in _layer.name:
        print(idx, "=>", _layer.name)
        conv_layers.append(idx)

print('conv_layers:', conv_layers)
print('layers_by_id:', layers_by_id)
```

```python
# image = cv2.resize(image, (96, 96))
# ValueError: Cannot feed value of shape (1, 3, 89, 109) for Tensor 'conv2d_1_input:0', which has shape '(?, 109, 89, 3)'
# 109x89 to 89x109 (cambia x por y e y por x)
image = cv2.resize(image, (89, 109))
image = img_to_array(image)
heatmaps = []
saliency = []

# Generates a gradient based class activation map 
# (grad-CAM) that maximizes the outputs of filter_indices in layer_idx.
# heat_map = visualize_cam(model, layer, None, image)

for idx_layer in conv_layers:
    # por algun motivo, no deja ver la primera capa convolucional
    heatmaps.append(cv2.resize(visualize_cam(model, idx_layer, None, image), (89*2, 109*2)))

# Generates an attention heatmap over the seed_input 
# for maximizing filter_indices output in the given layer_idx.
# heat_map_saliency = visualize_saliency(model, layer, None, seed_input=image)
for idx_layer in conv_layers:
    print(idx_layer)
    saliency.append(cv2.resize(visualize_saliency(model, idx_layer, None, seed_input=image), (89*2, 109*2)))
```

```python
# original image
destRGB = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
destRGB = cv2.resize(destRGB, (89*3, 109*3))

plt.figure(0)
plt.title("original image")
plt.imshow(destRGB)
plt.show()
```

```python
# heatmaps
for idx, image in zip(conv_layers[1:], heatmaps):
    plt.figure(idx)
    plt.title("grad-cam {}".format(layers_by_id[idx]))
    plt.imshow(image)
    destRGB = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2RGB)
    destRGB = cv2.resize(destRGB, (89*4, 109*4))
    cv2.imwrite("heatmaps/grad-cam/"+layers_by_id[idx]+".png", destRGB)
```

```python
# saliency
for idx, image in zip(conv_layers[1:], saliency):
    # primer filtro muestra una imagen toda roja, idk why
    plt.figure(idx)
    plt.title("saliency {}".format(layers_by_id[idx]))
    plt.imshow(image)
    destRGB = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2RGB)
    destRGB = cv2.resize(destRGB, (89*4, 109*4))
    cv2.imwrite("heatmaps/saliency/"+layers_by_id[idx]+".png", destRGB)
```
