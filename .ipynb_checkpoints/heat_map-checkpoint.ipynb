```python
from keras.models import load_model
from keras.layers.core import Dense
from keras.optimizers import SGD
from keras.applications import VGG16
import keras.backend as K
import numpy as np
import argparse
import imutils
import cv2
from vis.utils import utils
from keras import activations
```

```python
# Build the VGG16 network with ImageNet weights
# model = load_model('../male_female_cnn/santa_not_santa.model')
model = VGG16(weights='imagenet', include_top=True)


# Utility to search for layer index by name. 
# Alternatively we can specify this as -1 since it corresponds to the last layer.
layer_idx = utils.find_layer_idx(model, 'predictions')

# Swap softmax with linear
model.layers[layer_idx].activation = activations.linear
model = utils.apply_modifications(model)
```

```python
from vis.utils import utils
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (18, 6)

# img1 = utils.load_img('../male_female_cnn/examples/eye_right.tiff', target_size=(224, 224))
# img2 = utils.load_img('../male_female_cnn/examples/eye_left.tiff', target_size=(224, 224))

img1 = utils.load_img('../male_female_cnn/examples/dave.jpeg', target_size=(224, 224))
img2 = utils.load_img('../male_female_cnn/examples/f_china.jpg', target_size=(224, 224))

print(img1.shape)
print(img2.shape)

f, ax = plt.subplots(1, 2)
ax[0].imshow(img1)
ax[1].imshow(img2)
```

```python
from vis.visualization import visualize_saliency, overlay
from vis.utils import utils
from keras import activations

# Utility to search for layer index by name. 
# Alternatively we can specify this as -1 since it corresponds to the last layer.
layer_idx = utils.find_layer_idx(model, 'predictions')

print(model.summary())
```

```python
f, ax = plt.subplots(1, 2)
for i, img in enumerate([img1, img2]):
    # 949 is the imagenet index corresponding to `strawberry`
    # there are no eyes in imagenet clases
    grads = visualize_saliency(model, layer_idx, filter_indices=949, seed_input=img)
    
    # visualize grads as heatmap
    ax[i].imshow(grads, cmap='jet')
```

```python

import numpy as np
import matplotlib.cm as cm
from vis.visualization import visualize_cam

for modifier in [None, 'guided', 'relu']:
    plt.figure()
    f, ax = plt.subplots(1, 2)
    plt.suptitle("vanilla" if modifier is None else modifier)
    for i, img in enumerate([img1, img2]):    
        # 20 is the imagenet index corresponding to `ouzel`
        grads = visualize_cam(model, layer_idx, filter_indices=949, 
                              seed_input=img, backprop_modifier=modifier)        
        # Lets overlay the heatmap onto original image.    
        jet_heatmap = np.uint8(cm.jet(grads)[..., :3] * 255)
        
        print(jet_heatmap.shape) # por que weon, por que tiene otra dimension?
        print(img.shape)
        ax[i].imshow(overlay(jet_heatmap, img))
```
